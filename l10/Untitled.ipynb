```python
from abc import abstractmethod
from typing import Callable, Optional, List
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from tqdm import tqdm
```

```python
SimpleFloatFunc = Callable[[float], float]
CoordFunc = SimpleFloatFunc
TimeFunc = SimpleFloatFunc
TwoArgFunc = Callable[[float, float], float]
CoordAndTimeFunc = TwoArgFunc

class HeatEquation:
    def __init__(self, f: CoordAndTimeFunc = lambda x, t: 0,
                 left: TimeFunc = lambda t: 0,
                 right: TimeFunc = lambda t: 0,
                 first_state: CoordFunc = lambda x: 0,
                 a: float=1,
                 x_precision: int=100,
                 time_precision: int=100,
                 end_time: float=10,
                 kappa: float=1):
        self.f = f
        self.left = left
        self.right = right
        self.a = a
        self.k = kappa
        self.n = x_precision
        self.time_precision = time_precision
        self.end_time = end_time
        self.x_net, self.time_net = self.get_net(x_precision, time_precision)      
        self.h = self.x_net[1] - self.x_net[0]
        self.tau = self.time_net[1] - self.time_net[0]
        self.layers = [[first_state(x) for x in self.x_net]]
        self.solve()

        
    def get_net(self, x_prec, t_prec):
        return np.linspace(0, self.a, x_prec), np.linspace(0, self.end_time, t_prec)
        
    def solve(self):
        for time in tqdm(np.linspace(0, self.end_time, self.time_precision)[1:], leave=False):
            self.layers.append(self.get_next_layer(time))

    def plot(self):
        ax = sns.heatmap(np.array(self.layers).T)
        plt.show()
        
    @abstractmethod
    def generate_next_layer(self, time):
        ...
```

```python
class HeatEquationWithExplicitSchema(HeatEquation):
    
    def get_next_layer(self, time) -> List[float]:
        u_first = self.left(time)
        new_layer = [u_first]
        prev_layer = self.layers[-1]
        
        for i, x in list(enumerate(self.x_net))[1:-1]:
            u_c = prev_layer[i - 1] - 2 * prev_layer[i] + prev_layer[i + 1]
            u_n = self.tau * ((self.k / self.h ** 2) * u_c + self.f(x, time)) + prev_layer[i]
            new_layer.append(u_n)

        u_last = self.right(time)
        new_layer.append(u_last)
        return new_layer
```

```python
class HeatEquationWithNonExplicitSchema(HeatEquation):
    def __init__(self, f: CoordAndTimeFunc = lambda x, t: 0,
                 left: TimeFunc = lambda t: 0,
                 right: TimeFunc = lambda t: 0,
                 first_state: CoordFunc = lambda x: 0,
                 a: float = 1,
                 x_precision: int = 100,
                 time_precision: int = 100,
                 end_time: float = 10,
                 kappa: float = 1):
        self.S = None
        super().__init__(f, left, right, first_state, a, x_precision, time_precision, end_time, kappa)

    def get_S_matrix(self, n):
        a = - self.k / self.h ** 2
        b = 1 / self.tau + 2 * self.k / self.h ** 2

        first_row = np.zeros(n)
        first_row[0] = b
        first_row[1] = a
        S = [first_row]

        for i in range(n - 2):
            row = np.zeros(n)
            row[i] = row[i + 2] = a
            row[i + 1] = b
            S.append(row)

        last_row = np.zeros(n)
        last_row[n - 2] = a
        last_row[n - 1] = b
        S.append(last_row)

        return np.array(S)

    def get_next_layer(self, time):
        if self.S is None:
            self.S = self.get_S_matrix(self.n - 2)

        u_first = self.left(time)
        u_last = self.right(time)
        prev_layer = self.layers[-1]
        b = np.zeros(self.n - 2)
        koef = self.k / self.h ** 2
        
        b[0] = prev_layer[1] / self.tau + self.f(self.x_net[1], time + self.tau) + u_first * koef
        b[self.n - 3] = prev_layer[-2] / self.tau + self.f(self.x_net[-2], time + self.tau) + u_last * koef

        for i, x in list(enumerate(self.x_net))[2:-2]:
            b[i - 1] = prev_layer[i] / self.tau + self.f(x, time + self.tau)

        solution = np.linalg.solve(self.S, b)
        return [u_first, *solution, u_last]
```

```python
explicit_schema = HeatEquationWithExplicitSchema
right = lambda t: 20 + np.sin(t) * 10
left = lambda t: 30
f = lambda x, t: 0
initial = lambda x: 20
explicit_schema(left=left, right=right, kappa=0.1, x_precision=100, time_precision=1000, f=f, first_state=initial).plot()
```

```python
HeatEquationWithNonExplicitSchema(left=left, right=right, kappa=0.1, x_precision=100, time_precision=1000, f=f, first_state=initial).plot()
```
